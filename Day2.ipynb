Numpy

create arrays

```python
import numpy as np

a = np.array([10,20,30,40,50])
x = np.arange(101, 110)
y = np.zeros((4))
z = np.ones((4,3))

print(f"Array: {a}")
print(f"Range: {x}")
print(f"zeros: {y}")
print(f"ones: {z}")


```

Indexing and Slicing

```python
print(f"Index {a[2]}")
print(f"slicing {a[2:6]}")
```

Basic arithmatic operations

```python
print("a + 2:", a + 2)
print("a * 2:", a * 2)
```

Basic Statistics

```python
print("Square of a ", np.square(a))
print("mean of a ", np.mean(a))
print("Standerd deviation", np.std(a))
```

Shaping Array

```python
matrix = np.arange(1,17).reshape((4,4))
print(matrix)
print(matrix.T)
```

boolean Index

```python

mtx= np.arange(1,10)
print("numbers > 2", mtx [mtx > 2])
print(mtx>2)
```

PANDS

create a DataFrame

```python
import pandas as pd
data = {
    "name": ['Kavita', 'savita', 'geeta', 'Kavya','lavanya'],
    "age": [23,25,28,27,25],
    "salary":[10000,20000,40000,40000,30000],
    "department":['engineering', 'HR','HR', 'Engineering', 'sales']
}
dd = pd.DataFrame(data)
dd
```

Basic Operations

```python
print("First 4 rows")
dd.head(4)


print("columns")
dd.columns

print("data types")
dd.dtypes
```

Filtering

```python
print("people with age > 25")
dd[dd['age'] > 25]
```

Grouping

```python
grouped = dd.groupby('department')['salary'].mean()
print("Average Salary by Department:")
grouped
```

Sorting

```python
print(" people are sorted with salary")
dd.sort_values(by='salary')
```

📊 Matplotlib: Building Blocks of Visualizations

line plot

```python
import matplotlib.pyplot as plt
years =[2020, 2021, 2022, 2023,2025]
revenue =[100 ,220,330,400,500]

plt.plot(years, revenue, marker= 'o', linestyle='--', color='red')
plt.title("Company revenue over years")
plt.xlabel("years")
plt.ylabel("revenue")

plt.xticks(years)
plt.show()
```

Bar Chart

```python
department = dd['department'].value_counts()
department.plot(kind='bar', color='pink')
plt.title("Employee count by department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

Histogram

```python
age =dd['age']
plt.hist(age, bins=4, color='violet', edgecolor='black')
plt.title("Employee count by department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

Pie Chart

```python
dd.groupby('department').size().plot(kind='pie', autopct='%1.1f%%')
plt.title("department Proportion")
plt.ylabel("")
plt.show()
```

Seaborn: Statistical Data Visualization

```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
print("\nInfo about the tips dataset:")
tips.info()

print("\nDescription of the tips dataset:")
tips.describe(include='all')
```

Pairplot

```python
sns.pairplot(tips, hue="sex")
plt.show()
```

Heatmap of correlation

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Boxplot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```
